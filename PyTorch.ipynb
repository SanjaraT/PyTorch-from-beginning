#Mathematical Operations

```python
y = torch.rand(3,3)
y
```

```python
print("Add: ", y + 2)
print("Sub: ", y - 2)
print("Mul: ", y * 2)
print("Div: ", y / 2)
print("Int Div: ", (y*100)//2)
print("Mod: ", ((y*100)//2)%2)
print("Power: ", y**2)
```

```python
# Element wise
a = torch.rand(2,3)
b = torch.rand(2,3)

print("Add: ", a + b)
print("Sub: ", a - b)
print("Mul: ", a * b)
print("Div: ", a / b)
print("Mod: ", a % b)
print("Power", a**b)
```

```python
# Specific Range
torch.clamp(y, min = 0.5, max = 1)
```

#Reduction Operations

```python
e = torch.randint(size=(2,3), low =0, high = 10, dtype=torch.float32)
e
```

```python
print("Total: ", torch.sum(e))
print("Column Total: ", torch.sum(e, dim= 0))
print("Row Total: ", torch.sum(e, dim= 1))
print("Mean: ", torch.mean(e))
print("Mean of Columns: ", torch.mean(e, dim =0))
print("Median: ", torch.median(e))
print("Max: ", torch.max(e))
print("Min: ", torch.min(e))
print("Product: ", torch.prod(e))
print("Standard Deviation: ", torch.std(e))
print("Variance: ", torch.var(e))


```

single NN

```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)
```

```python
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)
```

```python
w
```

```python
b
```

```python
z = w*x + b
z
```

```python
y_pred = torch.sigmoid(z)
y_pred
```

```python
import torch.nn.functional as F

loss = F.binary_cross_entropy(y_pred, y)
loss
```

```python
loss.backward()
```

```python
print(w.grad)
print(b.grad)
```
